```python
import numpy as np

from matplotlib import pyplot as plt
```

```python
dt = .001
tau = .1
noise_tau = .002
w11, w22 = 1, 1
w21, w12 = 1, 1
dur = 6
bifurcate_time = .8
sgm = 5
threshold = 70
S = 576
eta = S * .3
R_initial = 32
```

```python
def update_noise(ti, V1, V2):
    if ti >= bifurcate_time/dt and ti % 5 == 0:
        V1_n = np.random.randn() * eta
        V2_n = np.random.randn() * eta
        new_V1 = V1_n + V1
        new_V2 = V2_n + V2
    else:
        new_V1 = V1
        new_V2 = V2
    return new_V1, new_V2
```

```python
noise_arr_V1 = np.zeros(6000)
noise_arr_V2 = np.zeros(6000)
for ti in range(1, int(dur/dt)):
    noise_arr_V1[ti], noise_arr_V2[ti]  = update_noise(ti, S, S)
plt.plot(noise_arr_V1)
plt.plot(noise_arr_V2)
plt.xlim(left=1000, right=2000)
plt.ylim(bottom=500)
plt.show()
```
