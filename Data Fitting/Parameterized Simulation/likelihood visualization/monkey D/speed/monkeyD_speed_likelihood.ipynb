```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
df = pd.read_excel("monkeyD_speed_v3_result.xlsx")
```

```python
alpha = df['a'].to_numpy()
beta = df['b'].to_numpy()
S = df['S'].to_numpy()
sigma = df['sgm'].to_numpy()
tauR = df['tr'].to_numpy()
tauD = df['td'].to_numpy()
tauG = df['tg'].to_numpy()
nLL = df['nll'].to_numpy()
```

```python
opt_alpha = 3.491279
opt_beta = 1.446276
opt_sigma = 23.049477
opt_S = 60.477674
opt_tauR = 0.006396
opt_tauG = 0.006256
opt_tauD = 0.078333
min_nLL = 10102.20011
```

```python
tmp_fig = plt.figure(figsize=(5,5))
tmp_ax = tmp_fig.add_subplot(projection="3d")

tmp_figg = tmp_ax.scatter3D(tauR, tauD, nLL, c=nLL)
tmp_ax.view_init(15, 75)
```

```python
norm = matplotlib.colors.Normalize(vmin=np.min(nLL), vmax=np.max(nLL))

c_m = matplotlib.cm.GnBu_r
s_m = matplotlib.cm.ScalarMappable(cmap=c_m, norm=norm)
```

```python
fig = plt.figure(figsize=(16, 9), tight_layout=True)
gs = GridSpec(2, 13, figure=fig)

ax1 = fig.add_subplot(gs[0, 1:4], projection='3d')
ax2 = fig.add_subplot(gs[0, 5:8], projection='3d')
ax3 = fig.add_subplot(gs[1, 1:4], projection='3d')
ax4 = fig.add_subplot(gs[1, 5:8], projection='3d')
ax5 = fig.add_subplot(gs[1, 9:12], projection='3d')

fig1 = ax1.scatter3D(alpha, beta, nLL, c=nLL, cmap='GnBu_r', alpha=.3)
ax1.view_init(5, 60)
ax1.set_xlabel("alpha")
ax1.set_ylabel("beta")
ax1.set_zlabel("nLL")
ax1.plot(opt_alpha, opt_beta, min_nLL, 'X', ms=10, c='maroon')

fig2 = ax2.scatter3D(S, sigma, nLL, c=nLL, cmap='GnBu_r', alpha=.3)
ax2.view_init(5, 75)
ax2.set_xlabel("S")
ax2.set_ylabel("sigma")
ax2.set_zlabel("nLL")
ax2.plot(opt_S, opt_sigma, min_nLL, 'X', ms=10, c='maroon')

fig3 = ax3.scatter3D(tauG, tauR, nLL, c=nLL, cmap='GnBu_r', alpha=.3)
ax3.view_init(5, 15)
ax3.set_xlabel("tauG")
ax3.set_ylabel("tauR")
ax3.set_zlabel("nLL")
ax3.plot(opt_tauG, opt_tauR, min_nLL, 'X', ms=10, c='maroon')

fig4 = ax4.scatter3D(tauG, tauD, nLL, c=nLL, cmap='GnBu_r', alpha=.3)
ax4.view_init(5, 75)
ax4.set_xlabel("tauG")
ax4.set_ylabel("tauD")
ax4.set_zlabel("nLL")
ax4.plot(opt_tauG, opt_tauD, min_nLL, 'X', ms=10, c='maroon')

fig5 = ax5.scatter3D(tauR, tauD, nLL, c=nLL, cmap='GnBu_r', alpha=.3)
ax5.view_init(5, 75)
ax5.set_xlabel("tauR")
ax5.set_ylabel("tauD")
ax5.set_zlabel("nLL")
ax5.plot(opt_tauR, opt_tauD, min_nLL, 'X', ms=10, c='maroon')

cbar_ax = fig.add_axes([.95, 0.25, 0.01, 0.5])
fig.colorbar(s_m, cax=cbar_ax)

plt.savefig("monkeyD_speed_v3_likelihood.pdf", format="pdf")
```
