```python
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
df = pd.read_excel("monkeyD_speed_v3_result.xlsx")
```

```python
alpha = df['a'].to_numpy()
beta = df['b'].to_numpy()
S = df['S'].to_numpy()
sigma = df['sgm'].to_numpy()
tauR = df['tr'].to_numpy()
tauD = df['td'].to_numpy()
tauG = df['tg'].to_numpy()
nLL = df['nll'].to_numpy()
```

```python
tmp_fig = plt.figure(figsize=(5,5))
tmp_ax = tmp_fig.add_subplot(projection="3d")

tmp_figg = tmp_ax.scatter3D(tauR, tauD, nLL, c=nLL)
tmp_ax.view_init(15, 75)
```

```python
fig = plt.figure(figsize=(14, 9), constrained_layout=True)
gs = GridSpec(2, 14, figure=fig)

ax1 = fig.add_subplot(gs[0, :6], projection='3d')
ax2 = fig.add_subplot(gs[0, 8:14], projection='3d')
ax3 = fig.add_subplot(gs[1, :4], projection='3d')
ax4 = fig.add_subplot(gs[1, 5:9], projection='3d')
ax5 = fig.add_subplot(gs[1, 10:], projection='3d')

fig1 = ax1.scatter3D(alpha, beta, nLL, c=nLL)
ax1.view_init(5, 60)
ax1.set_xlabel("alpha")
ax1.set_ylabel("beta")
ax1.set_zlabel("nLL")
fig.colorbar(fig1, ax=ax1, shrink=.75)

fig2 = ax2.scatter3D(S, sigma, nLL, c=nLL)
ax2.view_init(5, 75)
ax2.set_xlabel("S")
ax2.set_ylabel("sigma")
ax2.set_zlabel("nLL")
fig.colorbar(fig2, ax=ax2, shrink=.75)

fig3 = ax3.scatter3D(tauG, tauR, nLL, c=nLL)
ax3.view_init(5, 15)
ax3.set_xlabel("tauG")
ax3.set_ylabel("tauR")
ax3.set_zlabel("nLL")
fig.colorbar(fig3, ax=ax3, shrink=.75)

fig4 = ax4.scatter3D(tauG, tauD, nLL, c=nLL)
ax4.view_init(5, 75)
ax4.set_xlabel("tauG")
ax4.set_ylabel("tauD")
ax4.set_zlabel("nLL")
fig.colorbar(fig4, ax=ax4, shrink=.75)

fig5 = ax5.scatter3D(tauR, tauD, nLL, c=nLL)
ax5.view_init(5, 75)
ax5.set_xlabel("tauR")
ax5.set_ylabel("tauD")
ax5.set_zlabel("nLL")
fig.colorbar(fig5, ax=ax5, shrink=.75)

plt.savefig("monkeyD_speed_v3_likelihood.eps", format="eps")
```
